**Creating a model that will predict whether or not the user will click on an ad based off the features of that user**

```python
# importing all the essential libraries that I will be needing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ad = pd.read_csv('advertising.csv')
```

```python
ad.head()
```

```python
# Making a checkpoint to see if any data is missing that needs to be processed 
sns.heatmap(ad.isnull())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad[['Daily Time Spent on Site', 'Age','Area Income','Daily Internet Usage', 'Male']]
y = ad['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 42)
```

```python
# Training and fitting the Logistic Regression model
from sklearn.linear_model import LogisticRegression
```

```python
# Creating an instance of the Logistic regression class
log = LogisticRegression(max_iter = 1000)
```

```python
log.fit(X_train, y_train)
```

```python
predications = log.predict(X_test)
```

```python
# making the classification report to evaluate the model
from sklearn.metrics import classification_report
print(classification_report(y_test, predications))
```

From the above classification report we can understand that our model is giving 97% accurate results
